```python
df_pagerank = pd.read_csv('../data/df_pagerank_stand.csv')
df_pagerank = df_pagerank.drop(columns=['Unnamed: 0'])
df_pagerank['Class'] = cc['Class']



from scipy.stats import ttest_ind
pd.options.mode.chained_assignment = None

pagerank_norm_avg = []
pagerank_fraud_avg = []

tests_pagerank = []

for t in range(1, 48):
    if t == 1:
        temp = df_pagerank.iloc[ 0    :   (df_pagerank['1'].notna()[::-1].idxmax()+1)]

    if t >1:
        temp   = temp   = df_pagerank.iloc[ (df_pagerank[str(t-1)].notna()[::-1].idxmax()+1)    :   (df_pagerank[str(t)].notna()[::-1].idxmax()+1)]


    temp['pct_rank'] = temp[str(t)].rank(pct=True)


    temp_mean   = temp[['pct_rank','Class']].groupby(by='Class').mean()
    pagerank_norm_avg.append(temp_mean.iloc[0,0])
    if len(temp_mean) ==2:
        pagerank_fraud_avg.append(temp_mean.iloc[1,0])
    else:
        pagerank_fraud_avg.append(0)


    temp_fraud = temp[temp['Class'] == 1]
    temp_norm = temp[temp['Class'] == 0]

    tests_pagerank.append(ttest_ind(temp_fraud['pct_rank'], temp_norm['pct_rank'], nan_policy='omit'))


from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(10, 6), dpi=80)

plt.plot(pagerank_fraud_avg, label = 'fraud')
plt.plot(pagerank_norm_avg, label = 'normal')
plt.legend()
for t in range(0,47):
    if tests_pagerank[t][1] < 0.05:
        plt.axvspan(t-0.5, t+0.5, facecolor='b', alpha=0.1)
```
