```python
import pandas as pd
import numpy as np
import networkx as nx
import gc
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook


cc =  pd.read_csv("../data/creditcard.csv")
cc.Time = np.floor(cc['Time'] / (60*60))
```

# Pagerank code

```python
#personalized pagerank, alpha of 0.9, 1000 iterations

df_pagerank = pd.DataFrame(index=range(len(cc)),columns=range(0))
for t in tqdm_notebook(range(0,48)):

    #load in network
    g_part= nx.read_gpickle("../data/pickles2/myGraph" + str(t) + ".gpickle")

    #make dictionary with length of network
    dict_pers = dict.fromkeys(range(len(g_part)), 0)

    frauds = cc[(cc.Class == 1) & (cc.Time <= t)].index.tolist()
    total_frauds = len(cc[(cc.Class == 1) & (cc.Time <= t)])

    #go over frauds if they are in the dataset (aka within the length of the network)
    #if fraud: replace dict value with 1 over total fraud
    for k in frauds:
        if k < len(g_part):
            dict_pers[k] = 1/total_frauds

    ranking = nx.pagerank(g_part, alpha=0.9, personalization=dict_pers, max_iter=1000, tol=1e-06, nstart=None, weight='dists', dangling=None)
    rank_ordered =  dict(sorted(ranking.items(), key=lambda item: item[0]))


    df_pagerank.loc[:,t] = pd.Series(rank_ordered)

    del dict_pers
    del frauds
    del total_frauds
    del ranking
    del rank_ordered
    del g_part
    gc.collect()

df_pagerank.to_csv('../data/df_pagerank_pers_i1000a09_hnsw.csv')
```

```python
df_pagerank = pd.read_csv('../data/df_pagerank_pers_i10000a09t-1.csv')
df_pagerank = df_pagerank.drop(columns=['Unnamed: 0'])
df_pagerank['Class'] = cc['Class']

from scipy.stats import ttest_ind
pd.options.mode.chained_assignment = None

pagerank_norm_avg = []
pagerank_fraud_avg = []

tests_pagerank = []

for t in range(1, 48):
    if t == 1:
        temp = df_pagerank.iloc[ 0    :   (df_pagerank['1'].notna()[::-1].idxmax()+1)]

    if t >1:
        temp    = df_pagerank.iloc[ (df_pagerank[str(t-1)].notna()[::-1].idxmax()+1)    :   (df_pagerank[str(t)].notna()[::-1].idxmax()+1)]


    temp['pct_rank'] = temp[str(t)].rank(pct=True)


    temp_mean   = temp[['pct_rank','Class']].groupby(by='Class').mean()
    pagerank_norm_avg.append(temp_mean.iloc[0,0])
    if len(temp_mean) ==2:
        pagerank_fraud_avg.append(temp_mean.iloc[1,0])
    else:
        pagerank_fraud_avg.append(0)


    temp_fraud = temp[temp['Class'] == 1]
    temp_norm = temp[temp['Class'] == 0]

    tests_pagerank.append(ttest_ind(temp_fraud['pct_rank'], temp_norm['pct_rank'], nan_policy='omit'))


from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(10, 6), dpi=80)

plt.plot(pagerank_fraud_avg, label = 'fraud')
plt.plot(pagerank_norm_avg, label = 'normal')
plt.legend()
for t in range(0,47):
    if tests_pagerank[t][1] < 0.05:
        plt.axvspan(t-0.5, t+0.5, facecolor='b', alpha=0.1)
# # for t in range(1,28):
# #     if tests_pagerank[t][1] < 0.05:
# #         plt.vlines(t, 0, max(pagerank_norm_avg),colors='green', ls=':', lw=10)
# plt.title('alpha=0.85, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight=''dists'', dangling=None')
```
